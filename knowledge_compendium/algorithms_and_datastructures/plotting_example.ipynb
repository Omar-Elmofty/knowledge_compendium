```rust
:dep plotters = { version = "^0.3.6", default-features = false, features = ["evcxr", "all_series", "all_elements"] }
extern crate plotters;
use plotters::prelude::*;

// Define the drawing size
const OUT_FILE_NAME: &str = "evcxr_plot.png";
const WIDTH: u32 = 640;
const HEIGHT: u32 = 480;

let figure = evcxr_figure((WIDTH, HEIGHT), |root| {
    // 1. Create a chart context
    let mut chart = ChartBuilder::on(&root)
        .caption("Sine Wave Example", ("sans-serif", 30).into_font())
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(-3.14f32..3.14f32, -1.2f32..1.2f32)?; // Set X and Y range (-pi to pi, -1.2 to 1.2)

    // 2. Configure the mesh (grid and axes)
    chart.configure_mesh().draw()?;

    // 3. Draw the series (the actual line)
    chart.draw_series(LineSeries::new(
        // Generate points for the sine function: (x, sin(x))
        (-300..=300).map(|i| {
            let x = i as f32 / 100.0;
            (x, x.sin())
        }),
        &RED, // Use the color Red
    ))?.label("sin(x)").legend(|(x, y)| PathElement::new(vec![(x, y), (x + 20, y)], &RED));

    // 4. Configure the legend
    chart.configure_series_labels().border_style(&BLACK).draw()?;

    // 5. Return Ok to indicate success
    root.present()?;
    Ok(())
});
figure
```
