# Graphs



The bound on the number of edges of a fully connected graph
- lower_bound: $n-1$ or $O(n)$, imagine arranging all the nodes in a row, and connecting them all, this will create $n-1$ edges, any number less than that will cause at least one node to be not connected to the rest of the graph.
- upper_boubd: n choose 2 $\frac{n(n-1)}{2}$ or $O(n^2)$, each unique pair of nodes are connected by an edge.

Typically we refer to **sparse** graphs as graphs that have the number of edges closer to the lower bound, and **dense** graphs as  graphs where the number of edges are closer to the upper bound,


Typical graph notation:
- $G$: Graph
- $n$: number of nodes
- $m$: number of edges

## Breadth first search
explore the graph in layers.
If all the edges have equal cost, then BFS will arive to the shortest path
Runtime is O(n + m). Think about the FIFO queue that we use, it is simply n dimensional, with an additional m work for each edge.

Applications of breadth first search: shortest path (if edges have equal weights), inspecting the connectedness of graphs.

Proof as to why breadth first search finds the shortest path: BFS searches the graph by layers, peeling off each layer at a time. Assuming all edges have the same cost, then the shortest path will simply be the number of layers till we reach the goal, and since we peel the graph layer by layer from the start, then we will always find the shortest path (layer) to the goal.


The BFS algorithm could also be used to inspect the connectedness of the graph, or the number of islands in the graph.




## Depth first Search

If BFS is the cautious version of the search, DFS is it's more aggressive cousin.

DFS can be used to find the topological ordering (or topolgical sort) of a Directed Acyclic Graph (DAG).

It can also be used to compute the strongly connected components of a graph (due to the topological sort ability of this algorithm).
