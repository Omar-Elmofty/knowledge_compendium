# Diff Drive CLF CBF Formulation




```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 1000
dt = 0.1
c = 1.0      # CLF rate
alpha = 0.1   # CBF rate
delta_relaxation = 10.0

# Obstacle parameters
ox, oy, r = 25, 25, 5

def make_sdf(x, y):
    dx = x - ox
    dy = y - oy
    return np.sqrt(dx**2 + dy**2) - r

def grad_sdf(x, y):
    norm = np.sqrt((x - ox)**2 + (y - oy)**2)
    return np.array([(x - ox) / norm, (y - oy) / norm])

# Initial state
x, y, theta = 50.5, 50.0, -np.pi * 3 /4
xs, ys, thetas = [x], [y], [theta]
us = []
v_s = []
w_s = []
slack_s = []

for i in range(N):
    # c -= i  / N
    # Control input
    u = cp.Variable(3)

    x_dot_bl = u[0]
    y_dot_bl = 0.0
    if len(us) != 0:
        y_dot_bl = 0.5 * dt * (us[-1][0]*u[1] + us[-1][1] * u[0])
    x_dot = np.cos(theta) * x_dot_bl - np.sin(theta) * y_dot_bl
    y_dot = np.sin(theta) * x_dot_bl + np.cos(theta) * y_dot_bl

    # CLF constraint
    V = 0.5 * (x**2 + y**2)
    V_dot = (x * x_dot + y * y_dot)
    clf_constraint = V_dot + c * V <= u[2] #delta_relaxation

    # CBF constraint
    h = make_sdf(x, y)
    grad_h = grad_sdf(x, y)
    h_dot = grad_h[0] * x_dot + grad_h[1] * y_dot
    cbf_constraint = h_dot + alpha * h >= 0

    cap1 = u[0]**2 + u[1]**2 <= 4.0
    v_prev = 0
    w_prev = 0
    if len(us) != 0:
        v_prev = us[-1][0]
        w_prev = us[-1][1]
    cap2 = (u[0] - v_prev)**2 + (u[1] - w_prev)**2 <= dt * 0.5

    # Optimization problem
    cost = u[0]**2 + u[1]**2 + delta_relaxation * u[2]**2
    prob = cp.Problem(cp.Minimize(cost), [clf_constraint, cbf_constraint, cap1, cap2])
    prob.solve()

    # Apply control
    u_val = u.value
    us.append(u_val)
    v_s.append(u_val[0])
    w_s.append(u_val[1])
    slack_s.append(u_val[2])
    x += u_val[0] * np.cos(theta) * dt
    y += u_val[0] * np.sin(theta) * dt
    theta += u_val[1] * dt
    xs.append(x)
    ys.append(y)
    thetas.append(theta)

    if np.sqrt(x**2 + y**2) < 1:
        print("Goal Reached")
        break

# Plot the trajectory
xs = np.array(xs)
ys = np.array(ys)

circle = plt.Circle((ox, oy), r, color='r', alpha=0.3, label='Obstacle')
plt.figure(figsize=(6, 6))
plt.gca().add_patch(circle)
plt.plot(xs, ys, markersize=3, label="Trajectory")
plt.plot(xs[0], ys[0], 'go', label='Start')
plt.plot(xs[-1], ys[-1], 'rx', label='End')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D CLF-QP with CBF Obstacle Avoidance')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
